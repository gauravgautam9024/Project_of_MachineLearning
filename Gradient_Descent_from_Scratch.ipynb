```python
What is random variaiton ?
Random variation ka matlab hota hai woh natural fluctuations ya differences jo kisi system ya process mein chance ya randomness ki wajah se hote hain. Yeh variations kai factors ki wajah se ho sakte hain, jaise measurement errors, environmental conditions, ya system ke inherent randomness.

Example ke liye, agar aap ek fair coin ko 10 baar flip karte hain, toh har baar coin flip karne pe 50% chance hota hai heads aane ka aur 50% chance hota hai tails aane ka. Lekin, random variation ki wajah se, aapko har baar exactly 5 heads aur 5 tails nahi milenge. Ho sakta hai aapko kabhi 6 heads aur 4 tails mile, ya 4 heads aur 6 tails mile, ya kabhi kabhi 7 heads aur 3 tails bhi mil sakte hain. Yeh random fluctuations results mein random variation ka example hain.

Random variation ek important concept hai statistics aur data analysis mein kyunki yeh experiments ya studies ke results ko affect kar sakta hai. Random variation ko samajhkar aur account karke, hum data mein true underlying relationships ya patterns ko behtar tarike se estimate kar sakte hain aur more accurate conclusions bana sakte hain.


    
What is noise in data ?
Data analysis mein, noise ka matlab hota hai woh random variations ya errors jo data mein hoti hain aur jo underlying signal ya phenomenon se related nahi hoti hain jo measure kiya ja raha hai. Yeh random variations kai factors ki wajah se ho sakte hain, jaise measurement errors, data collection errors, ya natural variations in the data.

Noise ek problem ban sakta hai data analysis mein kyunki yeh true relationship ko variables ke beech obscure kar sakta hai ya data mein patterns ko detect karna mushkil bana sakta hai. Kuch cases mein, data cleaning ya filtering techniques ke through noise ko remove ya reduce karna possible hota hai. Dusre cases mein, noise ko analysis mein account karna zaroori hota hai, jiske liye statistical methods ya models use karte hain jo noise ke liye robust hote hain.

Machine learning mein, dataset mein noise add karna ek tarika hota hai real-world data ko simulate karne ka aur models ki generalization performance ko improve karne ka. Lekin, zyada noise add karna bhi mushkil bana sakta hai models ke liye data mein underlying patterns ko learn karna, isliye noise aur signal ke beech balance banana important hota hai.
```

```python
class GDRegressor : 
    
    def __init__(self , learning_rate , epochs):
        self.m = 10   # we can start any number as like m = 0 
        self.b = -1 # we can start any number as like b = 1 
        self.lr = learning_rate 
        self.epochs = epochs 
        
    def fit(self , X,y):
        #Calculate the b using GD 
        for i in range(self.epochs):
            loss_slope_b = -2*np.sum(y - self.m*X.ravel() - self.b)
            loss_slope_m = -2*np.sum((y - self.m*X.ravel() - self.b)*X.ravel())  
            
            self.b = self.b - (self.lr * loss_slope_b)
            self.m = self.m - (self.lr * loss_slope_m) 
        print(self.b , self.m)  
        
    def predict(self , X):
        return self.m * X + self.b 
        
 
```

```python
gd = GDRegressor(0.01 , 100)
gd.fit(X_train,y_train) 
# gd.predict(X)
y_pred = gd.predict(X_test) 
from sklearn.metrics import r2_score 
r2_score(y_test , y_pred)
```
