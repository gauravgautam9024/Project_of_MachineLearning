```python
What is Naive Bayes:-
                     Ek supervised machine learning algorithm hai.

                     Classification ke liye use hota hai.

                     Bayes Theorem pe based hai aur assume karta hai ki saare features independent hain.

                    “Naive” ka matlab simple assumption – features ek dusre pe depend nahi karte
```

```python
Kyu istemal karte hain:-
Fast aur simple algorithm hai

Text classification me best hai (spam detection, sentiment analysis)

Chhote dataset pe bhi kaam karta hai

Probabilistic output deta hai (chance ka estimation)

Kya jarurat hai:-
Jab aapko category ya class predict karni ho based on features

Jab text ya binary data ho

Jab speed aur efficiency chahiye
```

```python
Real World Example
Email Spam Detection:

Email me words aate hain: “win”, “money”, “free”

Naive Bayes predict karta hai: Ye email spam hai ya not spam

Ye e-commerce me bhi use hota hai:

“Agar user ne mobile kharida → accessory recommend karna”
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\Social_Network_Ads.csv")
df.head()
```

```python
X = df.drop(columns=["User ID","Purchased","Gender"])
y= df["Purchased"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
sc = StandardScaler()
```

```python
x_train_new = sc.fit_transform(X_train)
x_test_new = sc.fit_transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
classifier = GaussianNB()     # gaussian its work on multiplr fetures columns
                                # MultinomialNB its  work only text basis
                                    # BernoulliNB its work only binary
```

```python
classifier.fit(x_train_new,y_train)
```

```python
y_pred = classifier.predict(x_test_new)
y_pred
```

```python
cn = confusion_matrix(y_test,y_pred)
y_train.sum()
```

```python
Naive Bayes me “Naive” ka matlab
“Naive” ka matlab hota hai “simple assumption”.

Kya assume karta hai?
Naive Bayes ye assume karta hai ki saare features ek dusre se independent hain.

Matlab agar humare paas 2 features hain: Age aur Income

Naive Bayes maanta hai ki Age aur Income ek dusre par depend nahi karte, chahe real life me ye ho sakta hai.

Example: Spam Email
Email me words hain:

“win”, “money”, “free”

Naive Bayes assume karega:

P(“win” | Spam) aur P(“money” | Spam) independent hain

P(“win” AND “money” | Spam) = P(“win”|Spam) × P(“money”|Spam)

Ye simple banata hai calculation ko, warna har combination ke liye probability nikalni padegi jo bahut heavy ho jata hai.
```
