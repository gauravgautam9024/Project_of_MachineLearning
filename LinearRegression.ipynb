```python
class MeraLR: #mera constructor
    
    def __init__(self): # initializing 2 things
        self.m = None
        self.b = None
        
    def fit(self,X_train,y_train):  #kyunki ye ek machine learning algorithm h, isme 2 methods toh hone hi chaiye
        
        num = 0
        den = 0
        
        for i in range(X_train.shape[0]):
            
            num = num + ((X_train[i] - X_train.mean())*(y_train[i] - y_train.mean()))
            den = den + ((X_train[i] - X_train.mean())*(X_train[i] - X_train.mean()))
        
        self.m = num/den
        self.b = y_train.mean() - (self.m * X_train.mean())
        print(self.m)
        print(self.b)       
    
    def predict(self,X_test):
        
        print(X_test)
        
        return self.m * X_test + self.b
```
