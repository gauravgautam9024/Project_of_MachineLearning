# Apriori :-
                Apriori algorithm Machine Learning (ML) ka part hai, aur ML ke andar ye Unsupervised Learning mein aata hai.
                Apriori ek bahut hi popular Data Mining algorithm hai jo market basket analysis ke liye use hota hai.

                Simple shabdon mein kahein toh, ye algorithm ye pata lagata hai ki "Agar koi insaan Item A kharidta hai,
                toh uske Item B kharidne ke kitne chances hain?"

Apriori Kaam Kaise Karta Hai?
Ye algorithm teen main pillars par tika hai:

Support: Ye batata hai ki koi item kitni baar pure data mein aaya hai.

Confidence: Ye batata hai ki agar Item A liya gaya, toh kitni baar Item B bhi saath mein liya gaya.

Lift: Ye batata hai ki A aur B ka saath hona mahaz ek ittefaq (chance) hai ya unke beech sach mein koi mazboot rishta hai.

```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# 1. Sample Data (Transactions)
data = {'Milk': [1, 1, 0, 1],
        'Bread': [1, 1, 1, 0],
        'Butter': [0, 1, 1, 1]}
df = pd.DataFrame(data)

# 2. Apriori Apply Karein (Minimum 50% support wale items)
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

# 3. Rules Banayein
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
